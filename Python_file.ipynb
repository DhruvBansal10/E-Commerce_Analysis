## Connecting Python to Sql server 

```python
## Connecting SSMS with jupyter notebooks 
import urllib
import pyodbc
import pandas as pd
from sqlalchemy import create_engine

# --- CONFIGURATION ---
SERVER = r"DHRUV-ACERASPIR\SQLEXPRESS"
DATABASE = "OlistDB"

# --- CONNECTION STRING (Windows Authentication) ---
conn_str = (
    r"DRIVER={ODBC Driver 17 for SQL Server};"
    f"SERVER={SERVER};"
    f"DATABASE={DATABASE};"
    r"Trusted_Connection=yes;"
)

# Encode the connection string for SQLAlchemy
params = urllib.parse.quote_plus(conn_str)
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")


# --- REUSABLE FUNCTION TO EXPORT DATA ---
def export_to_sql(dataframe, table_name, if_exists_strategy="replace"):
    """Exports a pandas DataFrame to SQL Server."""
    try:
        dataframe.to_sql(
            name=table_name, con=engine, if_exists=if_exists_strategy, index=False
        )
        print(
            f"Successfully uploaded data to table '{table_name}' in '{DATABASE}'!"
        )
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
table_name = ["customers",'order_items','order_reviews',
              'orders','products','order_payments','sellers']
for i in range(7) :
    export_to_sql(dfs[i],table_name[i])
```
